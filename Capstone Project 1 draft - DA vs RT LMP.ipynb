BACKGROUND

Why is cost and profit important for Utilities and Merchants?

Although Utilities can pass through their cost to their customers, it is not ideal to go in front of the Public Service Commission (PSC) and request rates to be raised. Customers do not like to see their electric bill go up.  Tt is the best interest for Utilities to keep their cost low without raising the ire of the PSC who represents the customers interest. Utilties are allowed to raise funds from public or private organizations to maintian operations or capital investments. Shareholders expects a ROI from Utilities profit for lending the funds.    

Merchant are private company that acuqired public Utilities. PSC authorized conversion from public to private because the Merchant agreed a private organization will be able to cover cost via a private entity. Merchants cover cost by managing the generation and transmission portolio more efficently than a public utility. If they do not cover cost, a Merchant absorbs the cost which will reduce the amount of profit they have will receive. However, Shareholders may not receive their expected ROI.  

What is the Day-Ahead and Real-Time Market?

Utilities and Merchants participates in the Energy Market. The Day-Ahead Energy Market lets market participants commit to buy or sell wholesale electricity one day before the operating day, to help avoid price volatility. ... The Real-Time Energy Market balances the differences between day-ahead commitments and the actual real-time demand for and production of electricity.
(Source https://www.iso-ne.com/markets-operations/markets/da-rt-energy-markets/)
